```python
%load_ext sql
%config SqlMagic.style = '_DEPRECATED_DEFAULT'    
%sql sqlite:///dbsqlite.db
```

```python
tables = %sql SELECT name FROM sqlite_master WHERE type='table';

# Iterate over each table and print the first 20 rows
for table in tables:
    table_name = table[0]  # Access the table name using index 0
    print(f"Table: {table_name}")
    result = %sql SELECT * FROM {table_name} LIMIT 20;
    for row in result:
        print(row)
    print("\n")
```
